```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import statistics as st
import sympy as sy

def rdexcel(a):
    return pd.read_excel(a).as_matrix()

def plot3D(a,b):
    x,y=np.meshgrid(a,b);
    z=x**2+y**2; #write relation here
    plt.figure().add_subplot(111,projection='3d').plot_surface(x,y,z)

def syplot(eq,*argv):
    tup=tuple();
    for i in argv:
        tup=tup+tuple(i)
    sy.plotting.plot3d(eq,tup)   
```

```python
#LDA for 0 and 1 only

a=pd.read_csv("mnist_train.csv")
aa=np.array(a)
#print(aa.shape)
b=pd.read_excel('train_num.xlsx')
bb=np.array(b)
#print(b)
x0 = np.array([aa[i,:] for i in range(len(bb)) if bb[i,0]==0])
x0m=np.array([[np.mean(x0[:,i])] for i in range(784)])
#print(x0m.shape)
x1=np.array([aa[i,:] for i in range(len(bb)) if bb[i,0]==1])
x1m=np.array([[np.mean(x1[:,i])] for i in range(784)])
#print(x1.shape)
#print(x0.shape)
cov1=np.cov(x1.T)
cov0=np.cov(x0.T)
#print(cov1.shape,cov0.shape)
covv=cov1+cov0
cov=covv/2
U, s, V = np.linalg.svd(cov)
S = np.diag(s)
S=np.matrix(S)
Sinv=S.I
a1=pd.read_csv("mnist_test.csv")
aaa1=np.array(a1)
#print(aaa1.shape)
b1=pd.read_excel('test_num.xlsx')
bbb1=np.array(b1)
aa1=np.array([aaa1[i,:] for i in range(len(aaa1)) if bbb1[i,0]==0 or bbb1[i,0]==1])
bb1=np.array([bbb1[i,:] for i in range(len(aaa1)) if bbb1[i,0]==0 or bbb1[i,0]==1])
#print(bb1.shape)
count=0
comp1=np.sqrt(Sinv)*U.T
comp2=comp1*x0m
comp3=comp1*x1m
count0=0
count1=0
for i in range(len(aa1)):
    X=np.array([[aa1[i,j]] for j in range(784)])
    d1=(comp1*X - comp2).T*(comp1*X - comp2)
    d2=(comp1*X - comp3).T*(comp1*X - comp3)
    if(d1>=d2):
        if(bb1[i,0]==1):
            count+=1
            count0+=1
    else:
        if(bb1[i,0]==0):
            count+=1
            count1+=1
print('Correct Result',count,"/",len(aa1))
print('Accuracy: ',count/len(aa1)*100,'%')
```
