```python
import h5py as hp
import numpy as np

fil= hp.File('faceimages .mat','r')
data= np.array(fil['data'])
print(data.shape)
data=data[:10304,:]
print(data.shape)
cov=np.cov(data)
print(cov.shape)
eig_val,eig_vec =np.linalg.eig(cov)
no_of_eigen_vectors=1000
e_va=np.array(eig_val)
e_ve=np.array(eig_vec)

y=np.matmul(e_ve.T,data)
print(y.shape)
y=y[:no_of_eigen_vectors,:]
print(y.shape)
e_ve=e_ve[:,:no_of_eigen_vectors]
x=np.matmul(e_ve,y)
print(x.shape)
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
i=0
plt.figure()
im=list()
te=list()
dac=data
while(i<len(dac)):
    te.append(abs(dac[i][5]))
    if(len(te)==112):
        er=list()
        for j in range(0,len(te)):
            er.append(te[j])
        im.append(er)
        te.clear()
    i=i+1
im=np.array(im)
im=im.T
print(np.shape(im))
plt.imshow(im)
plt.figure()
i=0
im=list()
te=list()
dac=x
while(i<len(dac)):
    te.append(abs(dac[i][5]))
    if(len(te)==112):
        er=list()
        for j in range(0,len(te)):
            er.append(te[j])
        im.append(er)
        te.clear()
    i=i+1
im=np.array(im)
im=im.T
print(np.shape(im))
plt.imshow(im)
er=np.sqrt(np.sum(abs(data-x)**2))
print(er)
```
